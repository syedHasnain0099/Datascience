```python
#we import matplotlib to make plots for better 
import matplotlib.pyplot as plt
#we import numpy to create datasamples
import numpy as np

#we add this line because when we download our notebook file it will also download graph
%matplotlib inline
```

## Exapmle 1
## $f(x)=2x^2+4x+5$

```python
def f(x):
    return 2*x**2+4*x+5
```

```python
#make data
x_1=np.linspace(start=-5,stop=5,num=30)
x_1
```

```python
plt.title('graph of equation f(x)')

plt.xlim(-5,5)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)

plt.plot(x_1,f(x_1))
```

To reduce the cost of the function we will find slope and slope is find by taking the derivative of the fucntion

```python
def df(x):
    return 4*x+4
```

```python
#now we have to compare plots side by side
plt.figure(figsize=[15,5])
plt.subplot(1,2,1)
plt.title('graph of equation f(x)')
plt.xlim(-5,5)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1,f(x_1))

#2:plot


plt.subplot(1,2,2)
plt.title('graph of equation df(x)')
plt.xlabel('X',fontsize=16)
plt.grid()
plt.xlim(-4,5)
plt.ylim(-13,20)
plt.ylabel('df(x)',fontsize=16)
plt.plot(x_1,df(x_1))

```
